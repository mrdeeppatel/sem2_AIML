```python
# It plots charts from the data
import matplotlib.pyplot as mp

# X for x-axis
# It will be calculated if not given
x = [10,20,30,40,50,60,70]
# Y for y-axis
y = [20,20,30,10,50,40,20]


#changing default line
mp.plot(x,y)
mp.plot(x,y,'o')

#-It will include line also
mp.plot(x,y,marker='*')
mp.plot(x,y,'*')

#============================================
#Line Style
mp.plot(y,linestyle="-.")

mp.plot(y,linestyle=":")

mp.plot(y,"*",linestyle=":")

#==================================
# Color

mp.plot(y,"*",color="red")

mp.plot(y,"*",color="red",linestyle="-.")

#=====================================
# Size of marker and color

# ms = Marker Size
mp.plot(y,"*",ms=10,linestyle="-.",color="red")

# Marker edge color 
mp.plot(y,"*",ms=10,linestyle="-.",color="red",mec="b")

# Marker face color - color inside marker
mp.plot(y,"*",ms=20,linestyle="-.",color="red",mfc="b")

#=====================================
# Width of line
# lw = line width
mp.plot(y,"*",ms=10,linestyle="-.",color="red",mec='b',lw=5)

#=====================================
#Title and lable of graph

mp.plot(y)

#Title
mp.title("Numbers")

# X lable
mp.xlabel("Number X")

# Y lable
mp.ylabel("Number Y")

#=========================================
# Fonts 
# We have to create a dictionary to use fonts

font1={
    "family":'serif',
    "color":"b",
    "size":34
}
font2={
    "family":'serif',
    "color":"r",
    "size":24
}

mp.title("Numbers", fontdict=font1)

mp.xlabel("Number X", fontdict=font2)
mp.ylabel("Number Y", fontdict=font2)
#=========================================

# Ploting more then one data
y1=[23,32,34,12,32,45,25]
y2=[23,12,34,52,22,35,25]


# X will be calulated accordingly
mp.plot(y1)
mp.plot(y2)

# To pass X xsis also
mp.plot(x,y1,x,y2)

# =========================================

# Grid in graph 
mp.grid()

# =========================================

# Only one Axis
mp.grid(axis="y")

# =========================================
# Grid Color , style and width
mp.grid(color="b",linestyle=":",lw=1.4)

#=============================================

# Sub Plots
y1=[23,32,34,12,32,45,25]
y2=[23,12,34,52,22,35,25]
x=[1,2,3,4,5,6,7]


# #the figure has 1 row, 2 columns, and this plot is the first plot
mp.subplot(1,2,1)
mp.plot(x,y1)

# #the figure has 1 row, 2 columns, and this plot is the second plot.
mp.subplot(1,2,2)
mp.plot(x,y2)

# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

#the figure has 2 row, 1 columns, and this plot is the first plot
mp.subplot(1,2,1)
mp.title("Chart 1",color="b")
mp.plot(x,y1)

#the figure has 2 row, 1 columns, and this plot is the second plot.
mp.subplot(1,2,2)
mp.title("Chart 2",color="r")
mp.plot(x,y2)
mp.suptitle("Main Chart",size=20,color="yellow")
#=============================================


mp.show()
```
