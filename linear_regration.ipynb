```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import csv
import random

# Number of students
num_students = 20

# Generate student data
students = []
for i in range(1, num_students + 1):
    study_hours = round(random.uniform(1, 10), 2)  # Study hours between 1 and 10
    marks = round(study_hours * 10 + random.uniform(-5, 5), 2)  # Marks roughly correlated with study hours
    marks = max(0, min(100, marks))  # Ensure marks are between 0 and 100
    students.append({
        'StudentID': i,
        'StudyHours': study_hours,
        'Marks': marks
    })

# Define CSV file path
csv_file_path = 'student_data.csv'

# Write data to CSV
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['StudentID', 'StudyHours', 'Marks'])
    writer.writeheader()
    for student in students:
        writer.writerow(student)

print(f"Student data saved to {csv_file_path}")
```

```python
model =LinearRegression()
model.fit(X_train,Y_train)
```
