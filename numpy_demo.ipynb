```python
import numpy as np
# ARRAY
# 1D array

oneD=np.array([1,2,3,4,5,7])
print(oneD)
# printing Dimention of array
print(oneD.ndim)
# Shape of array
print(oneD.shape)
# Size of array
print(oneD.size)
# type 
print(oneD.dtype)
print("\n")

# 2D array

twoD=np.array([[1,2,3,4,5,6],[7,6,5,4,3,"2"]])
print(twoD)
# printing Dimention of array
print(twoD.ndim)
# Shape of array
print(twoD.shape)
# Size of array
print(twoD.size)
# Type
print(twoD.dtype)
```

```python
# Arithmetic operators
a=np.array([[1,2,3],[1,1,1]])
b=np.array([10,13,34])
c=np.array([10,14,44])

# both array should be same size
# it can be diffrent dymension

# 1st arithmatic operators 
# 2nd numpy function
print("Addition of array same D",c+b)
print("Addition of array same D",np.add(c,b))
print("\n")
print("substraction of array same D",c-b)
print("substraction of array same D",np.subtract(b,c))
print("\n")
print("multilpication of array same D",c*b)
print("multilpication of array same D",np.multiply(c,b))
print("\n")
print("divetion of array same D",c/b)
print("divetion of array same D",np.divide(c,b))
print("\n")
print("mod of array same D",c%b)
print("mod of array same D",np.mod(c,b))
print("\n")

#
print("Addition of array diffrent D",a+b)
```
