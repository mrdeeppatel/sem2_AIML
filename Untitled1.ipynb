```python
import pandas as pd

Product = pd.read_csv("products-1000.csv")
# print(Product)

# Drop Null values
new_product = Product.dropna()

# Fill Nul values
fill_na = Product.fillna(100)

#Fill Null in perticular column
perticula_col = Product["Category"].fillna("Makeup")

# Inplace True
Product["Category"].fillna("Makeup",inplace=True)

#Mean Median Mode
mean_price = Product["Price"].mean()
median_price = Product["Price"].median()
mode_price = Product["Price"].mode()
print(mean_price)
print(median_price)
print(mode_price)

```

```python
# Sorting
Product.sort_values(["Price"],ascending=0)
```

```python
# percentage
Product["Precentage"]=(Product["Stock"]*100)/Product["Price"]
print(Product["Precentage"])
```

```python
#Apply Function lambda FUnction
Product["Gread"] = Product["Price"].apply(
    lambda Price: "C" if 0 <= Price <= 400 else ("B" if 400 < Price <= 700 else "A")
)
print(Product["Gread"])

#Apply Function Normal Function
def categorize_price(price):
    if 0 <= price < 400:
        return "C"
    elif 400 < price <= 700:
        return "B"
    else:
        return "A"

# Apply the function to the 'Price' column
Product["Gread"] = Product["Price"].apply(categorize_price)
```
